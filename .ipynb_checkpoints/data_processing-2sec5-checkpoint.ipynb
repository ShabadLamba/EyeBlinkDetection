```python
import scipy.io
import numpy as np
from scipy.signal import find_peaks_cwt
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
mat = scipy.io.loadmat('EyeBlinks/2Blink4Seconds.mat')
print mat.key
```

```python
eyeblink_channel_1 = mat[1,:]
eyeblink_channel_2 = mat[2,:]
```

```python
print eyeblink_channel_1
indexes = find_peaks_cwt(eyeblink_channel_1, np.arange(1,40),min_length=5)
peaks = eyeblink_channel_1[indexes]
max_peak = peaks.max()
# print peaks
two_peaks = np.sort(peaks)[::-1][:3]
print two_peaks
print np.where(eyeblink_channel_1 == two_peaks[0])[0][0]
print np.where(eyeblink_channel_1 == two_peaks[2])[0][0]
```
