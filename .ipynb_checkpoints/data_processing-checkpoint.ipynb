```python
import scipy.io
from scipy.signal import find_peaks_cwt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
mat = scipy.io.loadmat('EyeBlinks/1Blink5Sec.mat')['Baseline_Segment_03']
```

```python
eyeblink_channel_1 = mat[1,:]
eyeblink_channel_2 = mat[2,:]
```

```python
print eyeblink_channel_1
```

```python
print eyeblink_channel_1.std()
print eyeblink_channel_1.mean()
print eyeblink_channel_1.var()
indexes = find_peaks_cwt(eyeblink_channel_1, np.arange(1,25))
peak = eyeblink_channel_1[indexes].max()
print peak
# max_value = eyeblink_channel_1.max()
# maxTermIndex = np.where(eyeblink_channel_1 == eyeblink_channel_1.max())
# peak = []
# print maxTermIndex
# for i in range(-2,3,1):
#     peak.append(eyeblink_channel_1[maxTermIndex[0][0]+i])
#print [i for i in eyeblink_channel_1 if i>(eyeblink_channel_1.std())]
```

```python
print eyeblink_channel_2.std()
print eyeblink_channel_2.mean()
print eyeblink_channel_2.var()
# max_value = eyeblink_channel_2.max()
# maxTermIndex = np.where(eyeblink_channel_2 == eyeblink_channel_2.max())
# peak_2 = []
# for i in range(-2,3,1):
#     peak_2.append(eyeblink_channel_2[maxTermIndex[0][0]+i])
indexes_2 = find_peaks_cwt(eyeblink_channel_2, np.arange(1,5))
peak_2 = eyeblink_channel_2[indexes_2].max()
print peak_2
#print [i for i in eyeblink_channel_2 if i>(eyeblink_channel_2.std())]
```

```python
from numpy.fft import fft, fftshift
window = np.hanning(abs(peak))
plt.plot(window)
plt.title("Hann window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
plt.show()
```

```python
plt.figure()
A = fft(window, 2048) / (len(window)/2.0)
freq = np.linspace(-0.5, 0.5, len(A))
response = 20 * np.log10(np.abs(fftshift(A / abs(A).max())))
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title("Frequency response of the Hann window")
plt.ylabel("Normalized magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
```

```python
plt.figure()
A = fft(window, 2048) / 25.5
mag = np.abs(fftshift(A))
freq = np.linspace(-0.5, 0.5, len(A))
response = 20 * np.log10(mag)
response = np.clip(response, -100, 100)
plt.plot(freq, response)
plt.title("Frequency response of the Hann window")
plt.ylabel("Magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
plt.axis('tight')
plt.show()
```
